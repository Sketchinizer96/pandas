####Combining Datasets: Merge and Join

```python
# The behaviour implemented in pd.merge() is a subset of what is kown as relational
#algebra, which is formal sets of rules for manipulating relational data,
# and forms the conceptual foundation of operations in databases
```

```python
# The pd.merge() function implements a number of types of joins: the one-to-one,
# many-to-one, and many-to-many joins. All three types of joins are accessed via an
# identical call to the pd.merge() interface; the type of join performed depends on the
# form of the input data.

# One to One Join :
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'hire_date': [2004, 2008, 2012, 2014]})
df3 = pd.merge(df1,df2)
df3
# The pd.merge() function recognizes that each DataFrame has an employee column,
# and automatically joins using this column as a key.
```

```python
# Many to One Join :
df4 = pd.DataFrame({'group':['Accounting','Engineering','HR'],
                    'supervisor':['Carly','Guido','Steve']})
pd.merge(df3,df4)
# The resulting DataFrame has an additional column with the supervisor information,
# where the information is repeated in one or more locations as required by the inputs.
```

```python
# Many to Many Join:
# If the key column in both the left and right array contains duplicates, then
# the result is a many-to-many merge.

df5 = pd.DataFrame({'group': ['Accounting', 'Accounting',
                              'Engineering', 'Engineering', 'HR', 'HR'],
                    'skills': ['math', 'spreadsheets', 'coding', 'linux',
                               'spreadsheets', 'organization']})
pd.merge(df1,df5)
```

```python
# Specification on merge key:
#This option works only if both the left and right DataFrames have the specified column name.
pd.merge(df1,df2,on='employee')
```

```python
# The left_on and the right_on:
# this works when the column names are different.
df3 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'salary': [70000, 80000, 120000, 90000]})
pd.merge(df1,df3,left_on='employee',right_on='name').drop('name',axis = 1)
# drop is used for dropping the redundant column.
```

```python
# the left_index and the right_index method
df1a = df1.set_index('employee')
df2a = df2.set_index('employee')
pd.merge(df1a,df2a, left_index=True,right_index= True)
```

```python
# For convenience, DataFrames implement the join() method, which performs a
# merge that defaults to joining on indices
df1a.join(df2a)
```

```python
# mix indices and columns :
pd.merge(df1a, df3, left_index=True, right_on='name')
```

```python
# Specifying Set Arithematic for Joins

df6 = pd.DataFrame({'name': ['Peter', 'Paul', 'Mary'],
                    'food': ['fish', 'beans', 'bread']},
                    columns=['name', 'food'])
df7 = pd.DataFrame({'name': ['Mary', 'Joseph'],
                    'drink': ['wine', 'beer']},
                    columns=['name', 'drink'])
pd.merge(df6,df7)
# mary was the only common name within the both data set
```

### GroupBy

```python
# The split step involves breaking up and grouping a DataFrame depending on the
# value of the specified key.
# The apply step involves computing some function, usually an aggregate, transformation,
# or filtering, within the individual groups.
# The combine step merges the results of these operations into an output array.
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                    'data': range(6)}, columns=['key', 'data'])
df
```

```python
df.groupby('key')
```

```python
# Aggregate,filter,transform,apply
rng = np.random.RandomState(0)
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data1': range(6),
                   'data2': rng.randint(0, 10, 6)},
                columns = ['key', 'data1', 'data2'])
df
```

```python
# Aggregation
df.groupby('key').aggregate(['min',np.median,max])
```

```python
df.groupby('key').aggregate({'data1':'min',
                              'data2': 'max'})
```

```python
# filter
def filter_func(x):
    return x['data2'].std()>4
print(df); print(df.groupby('key').std());
print(df.groupby('key').filter(filter_func))
# The filter() function should return a Boolean value specifying whether the group
# passes the filtering. Here because group A does not have a standard deviation greater
# than 4, it is dropped from the result.
```

```python
# Transformation. While aggregation must return a reduced version of the data, transformation
# can return some transformed version of the full data to recombine. For
# such a transformation, the output is the same shape as the input. A common example
# is to center the data by subtracting the group-wise mean
df.groupby('key').transform(lambda x:x -x.mean())
```

```python
# The apply() method. The apply() method lets you apply an arbitrary function to the
# group results. The function should take a DataFrame, and return either a Pandas
# object (e.g., DataFrame, Series) or a scalar; the combine operation will be tailored to
# the type of output returned.
def norm_by_data2(x):
    x['data1']/=x['data2'].sum()
    return x

df.groupby('key').apply(norm_by_data2)
```

```python
L = [0, 1, 0, 1, 2, 0]
print(df); print(df.groupby(L).sum())
```

```python
quartiles = np.percentile(births['births'],[25,50,75])
mu= quartiles[1]
sig = 0.74 * (quartiles[2]-quartiles[0])
# This final line is a robust estimate of the sample mean, where the 0.74 comes from the
# interquartile range of a Gaussian distribution.
```

```python
births = births.query('(births > @mu -5 *@sig) & (births < @mu + 5 * @sig)')
```

```python
# set 'day' column to integer; it originally was a string due to nulls
births['day'] = births['day'].astype(int)
```

```python
# create a datetime index from year,month,day
births.index=pd.to_datetime(10000 * births.year+
                            100 * births.month +
                            births.day, format = '%Y%m%d')
births['dayofweek'] = births.index.dayofweek
births
```

```python
from matplotlib import pyplot as plt
import matplotlib as mpl
births.pivot_table('births',index='dayofweek',
                  columns='decade', aggfunc='mean').plot()
plt.gca().set_xticklabels(['Mon','Tues','Wed','Thurs','Fri','Sat','Sun'])
plt.ylabel('mean births by day');
```

```python
%matplotlib inline
import seaborn;seaborn.set()

data.plot(figsize=(15,8))
plt.ylabel('Hourly Bicycle count');
```

```python
weekly = data.resample('w').sum()
weekly.plot(style=['-', '--', ':'],figsize=(15,8))
plt.ylabel('Weekly bicycle count');
```

```python
# Another way that comes in handy for aggregating the data is to use a rolling mean,
# utilizing the pd.rolling_mean() function. Here well do a 30-day rolling mean of our
# data, making sure to center the window
daily = data.resample('D').sum()
daily.rolling(30, center=True).sum().plot(style=['-', '--', ':'],figsize=(15,8))
plt.ylabel('mean hourly count');
```

```python
daily.rolling(50, center=True,win_type='gaussian').sum(std=10).plot(style=[':', '--', '-'],figsize=(15,8));
```

```python
by_time = data.groupby(data.index.time).mean()
hourly_ticks = 4 * 60 * 60 * np.arange(6)
by_time.plot(xticks=hourly_ticks, style=[':', '--', '-'],figsize=(15,8));
```

```python
by_weekday = data.groupby(data.index.dayofweek).mean()
by_weekday.index = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
by_weekday.plot(style=[':', '--', '-'],figsize=(15,8));
```

```python
weekend = np.where(data.index.weekday < 5, 'Weekday', 'Weekend')
by_time = data.groupby([weekend, data.index.time]).mean()
```
